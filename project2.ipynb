```python
numeric_columns = df.select_dtypes(include="number").columns

df_mean = df.copy()

df_mean[numeric_columns] = df_mean[numeric_columns].fillna(
    df_mean[numeric_columns].mean()
)

df_mean.isnull().sum()
```

```python
df_min = df.copy()

df_min[numeric_columns] = df_min[numeric_columns].fillna(
    df_min[numeric_columns].min()
)

df_min.isnull().sum()
```

```python
df_max = df.copy()

df_max[numeric_columns] = df_max[numeric_columns].fillna(
    df_max[numeric_columns].max()
)

df_max.isnull().sum()
```

```python
df_mean.to_csv("filled_beyza.csv", index=False)
```
